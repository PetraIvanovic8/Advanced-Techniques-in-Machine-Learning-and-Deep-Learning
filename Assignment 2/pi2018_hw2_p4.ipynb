# Problem 4 - Linear Separability

## 4.1

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the dataset
x_1_x = np.array([-1,1,-3,4])
x_1_y = np.array([-1,1,-3,4])

x_2_x = np.array([-1,1,-5,4])
x_2_y = np.array([1,-1,2,-8])

# Create a scatter plot to visualize the dataset
plt.scatter(x_1_x, x_1_y, label='Class 1', marker='o')
plt.scatter(x_2_x, x_2_y, label='Class 2', marker='x')

# Add labels and legend
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()

# Show the plot
plt.title("Original Dataset")
plt.show()
```

**Answer:**
This dataset is not linearly separable since we can not fit a linear line that would separate our data (x and o) in such a way where all the x's are on one side and o's on the other. Since the dataset is not linearly separable we cannot train a linear classifier to classify this dataset

## 4.2

```python
# Transformed data points
# z = x_1 * x_2
z1_x = x_1_x * x_1_y
z2_x = x_2_x * x_2_y

z_y = np.array([0,0,0,0])

# Create a scatter plot to visualize the dataset
plt.scatter(z1_x, z_y, label='Class 1', marker='o')
plt.scatter(z2_x, z_y, label='Class 2', marker='o')
plt.vlines(0, -.5, .5, label='Hyperplane', color='purple')
plt.scatter([0], [0], marker='x')
plt.title('Transformed data points')
plt.xlabel('z = x1*x2')
plt.ylabel ('y')
plt.legend()
plt.show()
```

**Answer:**

Yes, we can define a new 1-dimensional representation z in terms of x1 and x2 such that the dataset is
linearly separable in terms of 1-dimensional representation corresponding to z.

$ z = x_1 * x_2 $ as shown above.

## 4.3

**Answer:**

As we can see from the graphical representation above, the hyperline is a straight line centered at 0 - z=0. This shows us that non-linerly separable data can be separated after applying non-linear transformation as we did in our example of $ z = x_1 * x_2 $

**Answer:**

Nonlinear transformations can be used to make linearly inseparable datasets linearly separable.
